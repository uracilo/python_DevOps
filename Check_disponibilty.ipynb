```python
import pandas as pd
import requests
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
colnames=['site'] 
file = 'urls.txt'
URLS = pd.read_csv(file, names=colnames, header=None)
URLS.head(3)
```

```python
headers = {
    'User-Agent': 'Python_connection',
    'From': 'some_id' 
}
```

```python
def get_200 (url):
    try:
        response = requests.get(url,headers=headers)
        return response.status_code
    except requests.exceptions.ConnectionError as ConnectionError:
        return ConnectionError
    except requests.exceptions.HTTPError as HTTPError:
        return HTTPError
```

```python
def to_txt (url):
    string = str(url)
    return string
```

```python
URLS['response']=URLS.apply(lambda x: get_200(x['site']) ,axis=1)
```

```python
URLS['Response_service']=URLS.apply(lambda x: to_txt(x['response']) ,axis=1)
```

```python
URLS['Response_service_data'] = [x[:19] for x in URLS['Response_service']]
```

```python
response = URLS.pivot_table(index=['Response_service_data'], aggfunc='size')
```
